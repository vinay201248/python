# loc and iloc

```python
# loc will work with direct values and iloc wikl work with index values
```

```python
a = np.arange(1,61).reshape(10,6)
a
```

```python
b = pd.DataFrame(a,index=['i1','i2','i3','i4','i5','i6','i7','i8','i9','i10'],columns=['c1','c2','c3','c4','c5','c6'])
b
```

```python
b.loc['i1':'i3']
```

```python
b.loc['i1':'i3','c3']
```

```python
b.loc[['i1','i4'],['c3']]
```

```python
b.loc[['i2'],['c3']]
```

```python
# iloc with index
```

```python
b.iloc[0:4,0:3]
```

```python
b.iloc[0:3,2]
```
